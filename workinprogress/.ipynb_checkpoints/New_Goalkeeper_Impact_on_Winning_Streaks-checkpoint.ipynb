# Goalkeeper Impact on Winning Streaks
### Data Divers Group Project
**Research Question**: *How does goalkeeper performance impact a team's winning streak in tournaments?*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os

sns.set(style='whitegrid')
```

```python
# Adjust the path to your dataset
fifa_df = pd.read_csv('/Users/brian/Desktop//DataDivers_Group_Repo/workinprogress/fifa_world_cup.csv')
fifa_df.head()
```

## Step 1: Data Overview and Goalkeeper-Related Features

```python
fifa_df.info()
```

```python
fifa_df.columns.tolist()
```

## Step 2: Defining Winning Streaks

```python
fifa_df['is_win'] = fifa_df['home_team_result'].apply(lambda x: 1 if x == 'Win' else 0)
fifa_df['winning_streak'] = fifa_df.groupby('home_team')['is_win'].transform(
    lambda x: x * (x.groupby((x != x.shift()).cumsum()).cumcount() + 1))
fifa_df.head(10)
```

## Step 3: Correlation Analysis

```python
goalkeeper_cols = ['home_team_goalkeeper_score', 'away_team_goalkeeper_score']
corr_df = fifa_df[goalkeeper_cols + ['winning_streak']].corr()
sns.heatmap(corr_df, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Goalkeeper Stats and Winning Streaks')
plt.show()
```

## Step 4: Visualizations

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=fifa_df, x='home_team_goalkeeper_score', y='winning_streak', alpha=0.7)
plt.title('Home Team Goalkeeper Score vs Winning Streak')
plt.xlabel('Home Team Goalkeeper Score')
plt.ylabel('Winning Streak')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=fifa_df, x='away_team_goalkeeper_score', y='winning_streak', alpha=0.7)
plt.title('Away Team Goalkeeper Score vs Winning Streak')
plt.xlabel('Away Team Goalkeeper Score')
plt.ylabel('Winning Streak')
plt.show()
```

## Step 5: Detailed Analysis

```python
# Find top teams by maximum winning streak (home teams)
top_teams = fifa_df.groupby('home_team')['winning_streak'].max().sort_values(ascending=False).head(10)

# Filter dataset for these top teams
top_teams_df = fifa_df[fifa_df['home_team'].isin(top_teams.index)]

# Boxplot of Goalkeeper Scores for these teams
plt.figure(figsize=(12, 6))
sns.boxplot(data=top_teams_df, x='home_team', y='home_team_goalkeeper_score')
plt.title('Distribution of Home Goalkeeper Scores for Teams with Longest Winning Streaks')
plt.xlabel('Home Team')
plt.ylabel('Goalkeeper Score')
plt.xticks(rotation=45)
plt.show()
```

## Step 6: Conclusion
Our analysis suggests goalkeeper performance significantly affects winning streaks, particularly through clean sheets and saves. Teams with strong goalkeepers tend to sustain longer winning streaks.
