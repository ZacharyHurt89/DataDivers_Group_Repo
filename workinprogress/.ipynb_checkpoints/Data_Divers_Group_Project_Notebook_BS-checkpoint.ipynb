```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os
```

```python
#Count missing values in each column
missing_values = fifa_df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
# Exploring Question 1: "Can you predict what team is most likely to win the 2022 FIFA World Cup? 
# What are the key variables which explain which team has a higher chance of winning? 
# Can we calculate the expected winning rate for each team based on the history?"

# Step 1: Determine if the home team and away team won the match
fifa_df['home_win'] = fifa_df['home_team_score'] > fifa_df['away_team_score']
fifa_df['away_win'] = fifa_df['home_team_score'] < fifa_df['away_team_score']

# Step 2: Compute wins for home teams and away teams
home_wins = fifa_df.groupby('home_team')['home_win'].sum().reset_index()
away_wins = fifa_df.groupby('away_team')['away_win'].sum().reset_index()

# Step 3: Rename columns to a common team column name
home_wins.columns = ['team', 'home_wins']
away_wins.columns = ['team', 'away_wins']

# Step 4: Combine the wins into one DataFrame
wins = pd.merge(home_wins, away_wins, on='team', how='outer').fillna(0)
wins['total_wins'] = wins['home_wins'] + wins['away_wins']

# Step 5: Count total matches played by each team
teams_home = fifa_df[['home_team']].rename(columns={'home_team': 'team'})
teams_away = fifa_df[['away_team']].rename(columns={'away_team': 'team'})
teams_all = pd.concat([teams_home, teams_away])
total_matches = teams_all['team'].value_counts().reset_index()
total_matches.columns = ['team', 'total_matches']

# Step 6: Merge wins with total matches
team_performance = pd.merge(wins, total_matches, on='team', how='outer')
team_performance['win_rate'] = team_performance['total_wins'] / team_performance['total_matches']

# Step 7: Sort teams by win rate (highest first)
team_performance_sorted = team_performance.sort_values(by='win_rate', ascending=False)

print("Expected Winning Rates (based on historical match wins):")
print(team_performance_sorted[['team', 'total_wins', 'total_matches', 'win_rate']].head(10))

#Bar Plot: Top 10 Teams by Win Rate
# -------------------------------------
top10_win_rate = team_performance_sorted.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_win_rate, 
    x='win_rate', 
    y='team',
    color='skyblue'
)
plt.title('Top 10 Teams by Historical Win Rate')
plt.xlabel('Win Rate')
plt.ylabel('Team')
plt.tight_layout()
plt.show()

# 2. (Optional) Histogram of Win Rates for All Teams
# --------------------------------------------------
plt.figure(figsize=(8, 6))
sns.histplot(
    data=team_performance_sorted, 
    x='win_rate', 
    kde=True, 
    color='green'
)
plt.title('Distribution of Team Win Rates')
plt.xlabel('Win Rate')
plt.ylabel('Number of Teams')
plt.tight_layout()
plt.show()
```

```python
# Account for FIFA Rankings
# Step 1: Average FIFA rank for home and away matches
home_rank = fifa_df.groupby('home_team')['home_team_fifa_rank'].mean().reset_index()
away_rank = fifa_df.groupby('away_team')['away_team_fifa_rank'].mean().reset_index()

# Step 2: Rename for merging
home_rank.columns = ['team', 'avg_home_rank']
away_rank.columns = ['team', 'avg_away_rank']

# Step 3: Merge and compute overall average rank
avg_rank = pd.merge(home_rank, away_rank, on='team', how='outer')
avg_rank['avg_fifa_rank'] = (avg_rank['avg_home_rank'] + avg_rank['avg_away_rank']) / 2

print("Average FIFA Ranking for each team:")
print(avg_rank[['team', 'avg_fifa_rank']].sort_values(by='avg_fifa_rank'))

# Sort the avg_rank DataFrame by avg_fifa_rank (ascending: lower rank is better)
sorted_avg_rank = avg_rank.sort_values('avg_fifa_rank', ascending=True)

# Create a horizontal bar chart for the top 10 teams by average FIFA ranking
plt.figure(figsize=(12, 8))
sns.barplot(
    data=sorted_avg_rank.head(10),
    x='avg_fifa_rank',
    y='team',
    hue='team',
    dodge=False,        # Ensures bars don’t stack side-by-side
    palette='viridis',
    legend=False        # Hide the legend if each bar is a unique team anyway
)

plt.xlabel('Average FIFA Ranking')
plt.ylabel('Team')
plt.title('Top 10 Teams by Average FIFA Ranking (Lower is Better)')
plt.tight_layout()
plt.show()

```

```python
# Next construct a composite strenght index & provide visualizations for both win rate and composite scores
# Step 1: Create a composite strength index
# Since lower FIFA rank means a stronger team, we invert the ranking.
# For example, we can calculate: inverted_rank = (max_rank - avg_fifa_rank)
max_rank = avg_rank['avg_fifa_rank'].max()
avg_rank['inv_rank'] = max_rank - avg_rank['avg_fifa_rank']

# Step 2: Merge the performance metrics with the ranking information
combined = pd.merge(team_performance, avg_rank[['team', 'avg_fifa_rank', 'inv_rank']], on='team', how='outer')

# Step 3: Define a composite score: win_rate multiplied by the inverted rank.
combined['composite_score'] = combined['win_rate'] * combined['inv_rank']
combined_sorted = combined.sort_values(by='composite_score', ascending=False)
print("Top Teams by Composite Strength Score:")
print(combined_sorted[['team', 'win_rate', 'avg_fifa_rank', 'composite_score']].head(10))

# Visualization
# Step 1: Plot top 10 teams by win rate
plt.figure(figsize=(12, 6))
top10_win_rate = team_performance_sorted.head(10)
plt.bar(top10_win_rate['team'], top10_win_rate['win_rate'], color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Team')
plt.ylabel('Win Rate')
plt.title('Top 10 Teams by Historical Win Rate')
plt.tight_layout()
plt.show()

# Step 2: Plot top 10 teams by composite strength score
plt.figure(figsize=(12, 6))
top10_composite = combined_sorted.head(10)
plt.bar(top10_composite['team'], top10_composite['composite_score'], color='salmon')
plt.xticks(rotation=45)
plt.xlabel('Team')
plt.ylabel('Composite Strength Score')
plt.title('Top 10 Teams by Composite Strength Score')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def compute_avg_metric(home_metric, away_metric, new_col_name):
    # Compute the average for home matches
    home_avg = fifa_df.groupby('home_team')[home_metric].mean().reset_index().rename(
        columns={'home_team': 'team', home_metric: f'{new_col_name}_home'}
    )
    # Compute the average for away matches
    away_avg = fifa_df.groupby('away_team')[away_metric].mean().reset_index().rename(
        columns={'away_team': 'team', away_metric: f'{new_col_name}_away'}
    )
    # Merge the two averages
    merged = pd.merge(home_avg, away_avg, on='team', how='outer')
    # Calculate the overall average metric for the team
    merged[new_col_name] = merged[[f'{new_col_name}_home', f'{new_col_name}_away']].mean(axis=1)
    return merged[['team', new_col_name]]

# Calculate averages for each performance category:
gk_scores = compute_avg_metric(
    'home_team_goalkeeper_score', 
    'away_team_goalkeeper_score', 
    'gk_score'
)

defense_scores = compute_avg_metric(
    'home_team_mean_defense_score', 
    'away_team_mean_defense_score', 
    'defense_score'
)

midfield_scores = compute_avg_metric(
    'home_team_mean_midfield_score', 
    'away_team_mean_midfield_score', 
    'midfield_score'
)

offense_scores = compute_avg_metric(
    'home_team_mean_offense_score', 
    'away_team_mean_offense_score', 
    'offense_score'
)

# Merge all performance metrics into one DataFrame
performance = (
    gk_scores
    .merge(defense_scores, on='team', how='outer')
    .merge(midfield_scores, on='team', how='outer')
    .merge(offense_scores, on='team', how='outer')
)

# Calculate an overall performance score as the average of the four metrics
performance['overall_performance'] = performance[
    ['gk_score', 'defense_score', 'midfield_score', 'offense_score']
].mean(axis=1)

# Show the top 10 teams by overall performance
print(performance.sort_values('overall_performance', ascending=False).head(10))
```

```python
# Sort teams by overall_performance (descending) and select top 10
performance_sorted = performance.sort_values('overall_performance', ascending=False)
top10 = performance_sorted.head(10)

# 1. Bar Plot: Top 10 Teams by Overall Performance
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10, 
    x='team', 
    y='overall_performance', 
    color='skyblue'  # Single color to avoid palette warnings
)
plt.title('Top 10 Teams by Overall Performance')
plt.xlabel('Team')
plt.ylabel('Overall Performance Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Correlation Heatmap of the Performance Metrics
# We'll examine gk_score, defense_score, midfield_score, offense_score, and overall_performance
perf_cols = ['gk_score', 'defense_score', 'midfield_score', 'offense_score', 'overall_performance']
corr_matrix = performance[perf_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='coolwarm', 
    square=True
)
plt.title('Correlation Heatmap of Player Performance Metrics')
plt.tight_layout()
plt.show()
```

```python
# Using Player Performance Prediction by Ranking

# Merge the previous team_performance (which contains win_rate) with the performance DataFrame
combined = pd.merge(team_performance, performance[['team', 'overall_performance']], on='team', how='outer')

# Create a composite score (for example, a weighted sum of win_rate and overall_performance)
# You can adjust the weights based on what you think is more important.
combined['composite_score'] = (combined['win_rate'] * 0.5) + (combined['overall_performance'] * 0.5)

combined_sorted = combined.sort_values(by='composite_score', ascending=False)
print("Top Teams by Composite Score (Win Rate + Player Performance):")
print(combined_sorted[['team', 'win_rate', 'overall_performance', 'composite_score']].head(10))
```

```python
# 1. Bar Plot: Top 10 Teams by Composite Score
# --------------------------------------------
top10_composite = combined_sorted.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_composite, 
    x='team', 
    y='composite_score', 
    color='salmon'  # Single color for clarity
)
plt.title('Top 10 Teams by Composite Score (Win Rate + Player Performance)')
plt.xlabel('Team')
plt.ylabel('Composite Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Scatter Plot: Win Rate vs. Overall Performance
# -------------------------------------------------
# We can visualize how these two metrics relate, with the color (hue) 
# indicating the composite_score.

plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=combined, 
    x='win_rate', 
    y='overall_performance', 
    hue='composite_score',
    palette='viridis', 
    edgecolor='black'
)
plt.title('Win Rate vs. Overall Performance\nColored by Composite Score')
plt.xlabel('Win Rate')
plt.ylabel('Overall Performance')
plt.legend(title='Composite Score', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. (Optional) Correlation Heatmap of Win Rate, Overall Performance, and Composite Score
# ---------------------------------------------------------------------------------------
metrics_of_interest = ['win_rate', 'overall_performance', 'composite_score']
corr_matrix = combined[metrics_of_interest].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='coolwarm', 
    square=True
)
plt.title('Correlation Heatmap\n(Win Rate, Overall Performance, Composite Score)')
plt.tight_layout()
plt.show()
```
