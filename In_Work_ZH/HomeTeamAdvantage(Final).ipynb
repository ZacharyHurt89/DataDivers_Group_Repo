```python
print("\nData types:")
print(fifa_df.dtypes)
```

```python
print("\nMissing values in key columns:")
columns_of_interest = ['home_team', 'away_team', 'home_team_score', 'away_team_score', 
                       'neutral_location', 'home_team_result']
print(fifa_df[columns_of_interest].isnull().sum())
```

```python
if fifa_df['neutral_location'].dtype == object:
    print("\nConverting neutral_location from string to boolean...")
    fifa_df['neutral_location'] = fifa_df['neutral_location'].map({'True': True, 'False': False})
```

```python
print("\nUnique values for neutral_location:", fifa_df['neutral_location'].unique())
print("Unique values for home_team_result:", fifa_df['home_team_result'].unique())
```

```python
home_venue_matches = fifa_df[fifa_df['neutral_location'] == False]
neutral_venue_matches = fifa_df[fifa_df['neutral_location'] == True]
```

```python
print(f"\nTotal matches: {len(fifa_df)}")
print(f"Matches where home team plays at home venue: {len(home_venue_matches)} ({len(home_venue_matches)/len(fifa_df)*100:.2f}%)")
print(f"Matches played at neutral venues: {len(neutral_venue_matches)} ({len(neutral_venue_matches)/len(fifa_df)*100:.2f}%)")
```

```python
if 'Lose' in fifa_df['home_team_result'].unique() and 'Loss' not in fifa_df['home_team_result'].unique():
    # Standardize to "Loss" instead of "Lose"
    fifa_df['home_team_result'] = fifa_df['home_team_result'].replace('Lose', 'Loss')

print("\nCounts of match results by venue type:")
print(pd.crosstab(fifa_df['neutral_location'], fifa_df['home_team_result']))
```

```python
results_by_venue = pd.crosstab(fifa_df['neutral_location'], fifa_df['home_team_result'], normalize='index') * 100

print("\nPercentages of match results by venue type:")
print(results_by_venue)
```

```python
results_by_venue.index = ['Home Team at Home Venue', 'Neutral Venue']
```

```python
plt.figure(figsize=(14, 8))
results_by_venue.plot(kind='bar', figsize=(14, 8))
plt.title('Match Outcomes: Home Team at Home Venue vs. Neutral Venue', fontsize=16)
plt.xlabel('Venue Type', fontsize=14)
plt.ylabel('Percentage of Matches (%)', fontsize=14)
plt.legend(title='Match Result', fontsize=12)

for container in plt.gca().containers:
    plt.gca().bar_label(container, fmt='%.1f%%', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
contingency_table = pd.crosstab(fifa_df['neutral_location'], fifa_df['home_team_result'])
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nChi-square test for independence:")
print(f"Chi-square value: {chi2:.2f}")
print(f"P-value: {p_value:.6f}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies:\n{pd.DataFrame(expected, index=['Home Team at Home Venue', 'Neutral Venue'], columns=contingency_table.columns)}")
print(f"Statistically significant difference at α=0.05: {p_value < 0.05}")
```

```python
home_team_goals_at_home = home_venue_matches['home_team_score'].mean()
away_team_goals_at_home_venue = home_venue_matches['away_team_score'].mean()
home_team_goals_at_neutral = neutral_venue_matches['home_team_score'].mean()
away_team_goals_at_neutral = neutral_venue_matches['away_team_score'].mean()
```

```python
goal_diff_at_home = home_team_goals_at_home - away_team_goals_at_home_venue
goal_diff_at_neutral = home_team_goals_at_neutral - away_team_goals_at_neutral
```

```python
goal_df = pd.DataFrame({
    'Venue Type': ['Home Team at Home Venue', 'Neutral Venue'],
    'Home Team Avg Goals': [home_team_goals_at_home, home_team_goals_at_neutral],
    'Away Team Avg Goals': [away_team_goals_at_home_venue, away_team_goals_at_neutral],
    'Goal Difference (Home-Away)': [goal_diff_at_home, goal_diff_at_neutral]
})

print("\nGoal statistics comparison:")
print(goal_df)
```

```python
plt.figure(figsize=(14, 8))
x = np.arange(len(goal_df))
width = 0.25

plt.bar(x - width, goal_df['Home Team Avg Goals'], width, label='Home Team Avg Goals')
plt.bar(x, goal_df['Away Team Avg Goals'], width, label='Away Team Avg Goals')
plt.bar(x + width, goal_df['Goal Difference (Home-Away)'], width, label='Goal Difference')

plt.title('Goal Statistics by Venue Type', fontsize=16)
plt.ylabel('Average Goals', fontsize=14)
plt.xticks(x, goal_df['Venue Type'], fontsize=14)
plt.legend(fontsize=12)

for i, col in enumerate(['Home Team Avg Goals', 'Away Team Avg Goals', 'Goal Difference (Home-Away)']):
    for j, val in enumerate(goal_df[col]):
        plt.text(j + (i-1)*width, val + 0.05, f'{val:.2f}', ha='center', fontsize=10)

plt.tight_layout()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
goal_diffs_at_home = home_venue_matches['home_team_score'] - home_venue_matches['away_team_score']
goal_diffs_at_neutral = neutral_venue_matches['home_team_score'] - neutral_venue_matches['away_team_score']

t_stat, p_value_ttest = stats.ttest_ind(
    goal_diffs_at_home,
    goal_diffs_at_neutral,
    equal_var=False  # Welch's t-test for unequal variances
)

print(f"\nT-test for goal difference:")
print(f"t-statistic: {t_stat:.2f}")
print(f"P-value: {p_value_ttest:.6f}")
print(f"Statistically significant difference at α=0.05: {p_value_ttest < 0.05}")
```

```python
print("\nTournament types in the dataset:")
tournament_counts = fifa_df['tournament'].value_counts()
print(tournament_counts[tournament_counts > 100])  # Only show tournaments with >100 matches

tournament_home_advantage = []

for tournament in fifa_df['tournament'].value_counts().index:
    tournament_data = fifa_df[fifa_df['tournament'] == tournament]
    if len(tournament_data) < 100:  # Skip tournaments with few matches
        continue
        
    t_home_venue = tournament_data[tournament_data['neutral_location'] == False]
    t_neutral_venue = tournament_data[tournament_data['neutral_location'] == True]
    if len(t_home_venue) > 30 and len(t_neutral_venue) > 30:
        t_home_win_rate = t_home_venue[t_home_venue['home_team_result'] == 'Win'].shape[0] / len(t_home_venue) * 100
        t_neutral_win_rate = t_neutral_venue[t_neutral_venue['home_team_result'] == 'Win'].shape[0] / len(t_neutral_venue) * 100
        
        tournament_home_advantage.append({
            'Tournament': tournament,
            'Home Venue Matches': len(t_home_venue),
            'Neutral Venue Matches': len(t_neutral_venue),
            'Home Team Win Rate at Home Venue (%)': t_home_win_rate,
            'Home Team Win Rate at Neutral Venue (%)': t_neutral_win_rate,
            'Home Advantage (pp)': t_home_win_rate - t_neutral_win_rate
        })

```

```python
fifa_df['rank_diff'] = fifa_df['away_team_fifa_rank'] - fifa_df['home_team_fifa_rank']
```

```python
ranked_matches = fifa_df.dropna(subset=['rank_diff'])
```

```python
bins = [-100, -50, -20, -5, 0, 5, 20, 50, 100]
labels = ['-100 to -50', '-50 to -20', '-20 to -5', '-5 to 0', '0 to 5', '5 to 20', '20 to 50', '50 to 100']
ranked_matches['rank_diff_group'] = pd.cut(ranked_matches['rank_diff'], bins, labels=labels)
```

```python
rank_advantage = []
```

```python
for group in ranked_matches['rank_diff_group'].unique():
    group_data = ranked_matches[ranked_matches['rank_diff_group'] == group]
    
    g_home_venue = group_data[group_data['neutral_location'] == False]
    g_neutral_venue = group_data[group_data['neutral_location'] == True]
    
    # Only include groups with sufficient matches
    if len(g_home_venue) > 20 and len(g_neutral_venue) > 20:
        g_home_win_rate = g_home_venue[g_home_venue['home_team_result'] == 'Win'].shape[0] / len(g_home_venue) * 100
        g_neutral_win_rate = g_neutral_venue[g_neutral_venue['home_team_result'] == 'Win'].shape[0] / len(g_neutral_venue) * 100
        
        rank_advantage.append({
            'Rank Difference': group,
            'Home Venue Matches': len(g_home_venue),
            'Neutral Venue Matches': len(g_neutral_venue),
            'Home Team Win Rate at Home Venue (%)': g_home_win_rate,
            'Home Team Win Rate at Neutral Venue (%)': g_neutral_win_rate,
            'Home Advantage (pp)': g_home_win_rate - g_neutral_win_rate
        })

rank_advantage_df = pd.DataFrame(rank_advantage)

if not rank_advantage_df.empty:
    # Sort by rank difference for better visualization
    rank_order = {label: i for i, label in enumerate(labels)}
    rank_advantage_df['Order'] = rank_advantage_df['Rank Difference'].map(rank_order)
    rank_advantage_df = rank_advantage_df.sort_values('Order')
    
    print("\nHome advantage by FIFA ranking difference:")
    print(rank_advantage_df[['Rank Difference', 'Home Venue Matches', 'Neutral Venue Matches', 
                       'Home Team Win Rate at Home Venue (%)', 'Home Team Win Rate at Neutral Venue (%)', 'Home Advantage (pp)']])
    plt.figure(figsize=(14, 8))
    
    plt.plot(rank_advantage_df['Rank Difference'], rank_advantage_df['Home Team Win Rate at Home Venue (%)'], 
             'o-', linewidth=2, markersize=10, label='Home Team at Home Venue')
    plt.plot(rank_advantage_df['Rank Difference'], rank_advantage_df['Home Team Win Rate at Neutral Venue (%)'], 
             's-', linewidth=2, markersize=10, label='Home Team at Neutral Venue')
    
    plt.title('Win Rates by FIFA Ranking Difference', fontsize=16)
    plt.xlabel('FIFA Ranking Difference (Away - Home)', fontsize=14)
    plt.ylabel('Win Rate (%)', fontsize=14)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("\nInsufficient data to calculate home advantage by team strength")
```

```python
print("\n=== SUMMARY OF FINDINGS ===")
home_team_win_at_home = home_venue_matches[home_venue_matches['home_team_result'] == 'Win'].shape[0] / len(home_venue_matches) * 100
home_team_win_at_neutral = neutral_venue_matches[neutral_venue_matches['home_team_result'] == 'Win'].shape[0] / len(neutral_venue_matches) * 100

print(f"1. Home teams win {home_team_win_at_home:.1f}% of matches at their home venue compared to {home_team_win_at_neutral:.1f}% at neutral venues.")
print(f"2. The home advantage is {home_team_win_at_home - home_team_win_at_neutral:.1f} percentage points.")
print(f"3. Statistical significance: Chi-square p-value = {p_value:.6f}, {'significant' if p_value < 0.05 else 'not significant'} at α=0.05")
print(f"4. Home teams score on average {home_team_goals_at_home:.2f} goals at home vs {home_team_goals_at_neutral:.2f} at neutral venues.")
print(f"5. Away teams score on average {away_team_goals_at_home_venue:.2f} goals at the home team's venue vs {away_team_goals_at_neutral:.2f} at neutral venues.")
print(f"6. The goal difference advantage is {goal_diff_at_home - goal_diff_at_neutral:.2f} goals.")

# Final conclusion
print("\nCONCLUSION:")
if home_team_win_at_home > home_team_win_at_neutral and p_value < 0.05:
    print("The data provides strong evidence that there is indeed a home team advantage in international football.")
    print(f"Teams playing at their home venue have a {home_team_win_at_home - home_team_win_at_neutral:.1f} percentage point higher win rate compared to playing at neutral venues.")
    print("This advantage is statistically significant.")
elif home_team_win_at_home > home_team_win_at_neutral:
    print("The data suggests there may be a home team advantage, but the evidence is not statistically significant.")
    print(f"Teams playing at their home venue have a {home_team_win_at_home - home_team_win_at_neutral:.1f} percentage point higher win rate compared to playing at neutral venues.")
else:
    print("The data does not provide evidence of a home team advantage in international football.")
    print(f"Teams playing at their home venue actually have a {home_team_win_at_neutral - home_team_win_at_home:.1f} percentage point lower win rate compared to playing at neutral venues.")
```
