```python
#Question 5: "What team has the longest winning streak?"
# Visualize Result

fifa_ds['home_win'] = fifa_ds['home_team_score'] > fifa_ds['away_team_score']
fifa_ds['away_win'] = fifa_ds['home_team_score'] < fifa_ds['away_team_score']

home_wins = fifa_ds.groupby('home_team')['home_win'].sum().reset_index()
away_wins = fifa_ds.groupby('away_team')['away_win'].sum().reset_index()

home_wins.columns = ['team', 'home_wins']
away_wins.columns = ['team', 'away_wins']

wins = pd.merge(home_wins, away_wins, on='team', how='outer').fillna(0)
wins['total_wins'] = wins['home_wins'] + wins['away_wins']

teams_home = fifa_ds[['home_team']].rename(columns={'home_team': 'team'})
teams_away = fifa_ds[['away_team']].rename(columns={'away_team': 'team'})
teams_all = pd.concat([teams_home, teams_away])
total_matches = teams_all['team'].value_counts().reset_index()
total_matches.columns = ['team', 'total_matches']

team_performance = pd.merge(wins, total_matches, on='team', how='outer')
team_performance['win_rate'] = team_performance['total_wins'] / team_performance['total_matches']

team_performance_sorted = team_performance.sort_values(by='win_rate', ascending=False)

print("Expected Winning Rates (based on historical match wins):")
print(team_performance_sorted[['team', 'total_wins', 'total_matches', 'win_rate']].head(10))

#Bar Plot: Top 10 Teams by Win Rate
# -------------------------------------
top10_win_rate = team_performance_sorted.head(10)

plt.figure(figsize=(8, 6))
sns.barplot(
    data=top10_win_rate, 
    x='win_rate', 
    y='team',
    color='purple'
)
plt.title('Top 10 Teams by Historical Win Rate')
plt.xlabel('Win Rate')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
fifa_ds['date'] = pd.to_datetime(fifa_ds['date'])
fifa_ds = fifa_ds.sort_values(by='date')  # ensure chronological order
```

```python
def get_winner(row):
    if row['home_team_result'] == 'Win':
        return row['home_team']
    elif row['home_team_result'] == 'Lose':
        return row['away_team']
    else:
        return None  # Draw or no result

fifa_ds["winner"] = fifa_ds.apply(get_winner, axis=1)
```

```python
from collections import defaultdict

streaks = defaultdict(int)
max_streaks = defaultdict(int)
last_winner = None
current_streak_team = None
current_streak_count = 0
streak_history = []  # to record streak timeline for visualization

for _, row in fifa_ds.iterrows():
    winner = row["winner"]
    if winner == current_streak_team:
        current_streak_count += 1
    else:
        if current_streak_team:
            streaks[current_streak_team] = max(streaks[current_streak_team], current_streak_count)
        current_streak_team = winner
        current_streak_count = 1 if winner else 0

    streak_history.append({
        "date": row["date"],
        "team": current_streak_team,
        "streak_length": current_streak_count
    })

# One last update
if current_streak_team:
    streaks[current_streak_team] = max(streaks[current_streak_team], current_streak_count)

# Get the team with the longest streak
longest_streak_team = max(streaks, key=streaks.get)
longest_streak_length = streaks[longest_streak_team]
```

```python
import matplotlib.pyplot as plt

team_streak_df = pd.DataFrame(streak_history)
team_streak_df = team_streak_df[team_streak_df["team"] == longest_streak_team]

plt.figure(figsize=(10,6))
plt.plot(team_streak_df["date"], team_streak_df["streak_length"], marker="o")
plt.title(f"Longest Winning Streak - {longest_streak_team} ({longest_streak_length} wins)")
plt.xlabel("Date")
plt.ylabel("Streak Length")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_streaks = sorted(streaks.items(), key=lambda x: x[1], reverse=True)[:10]
teams, lengths = zip(*top_streaks)

plt.figure(figsize=(10,6))
plt.barh(teams, lengths)
plt.title("Top 10 Longest Winning Streaks by Team")
plt.xlabel("Number of Consecutive Wins")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
teams = fifa_ds['home_team'].unique()
team_streaks = []
longest_streaks = {}

for team in teams:
    team_matches = fifa_ds[(fifa_ds['home_team'] == team) | (fifa_ds['away_team'] == team)].sort_values('date')
    
    streak = 0
    max_streak = 0
    streak_dates = []
    
    for _, row in team_matches.iterrows():
        # Determine if this team won
        if row['home_team'] == team and row['home_team_result'] == 'Win':
            streak += 1
        elif row['away_team'] == team and row['home_team_result'] == 'Lose':
            streak += 1
        else:
            streak = 0  # Reset on draw or loss

        max_streak = max(max_streak, streak)
        team_streaks.append({'team': team, 'date': row['date'], 'streak': streak})
    
    longest_streaks[team] = max_streak

# Find team with longest streak
best_team = max(longest_streaks, key=longest_streaks.get)
best_streak = longest_streaks[best_team]
print(f"{best_team} has the longest streak: {best_streak} wins")
```
